```python
import pandas as pd

n = int(input("Enter a value greater than 0: "))
if n <= 0:
    print("Invalid input. Please enter a value greater than 0.")
else:
    even_numbers = [2 * i for i in range(n)]
    odd_numbers = [2 * i + 1 for i in range(n)]
    fibonacci_numbers = [0, 1]
    for i in range(2, n):
        fibonacci_numbers.append(fibonacci_numbers[i-1] + fibonacci_numbers[i-2])
    
    data = {"Even Numbers": even_numbers, "Odd Numbers": odd_numbers, "Fibonacci Numbers": fibonacci_numbers}
    df = pd.DataFrame(data)
    print(df)
```
